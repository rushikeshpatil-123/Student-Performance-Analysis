```python
# Data Manipulation Libraries
import numpy as np
import pandas as pd

# Data Visualization Libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Filter out Warning Messages
import warnings
warnings.filterwarnings(action = 'ignore')

```

```python

# Load Dataset
df = pd.read_csv("/content/StudentsPerformance.csv")

# Preview Data
print("First five rows")
print(df.head())

# Data Validation
print("\n Missing Values:")
print(df.isnull().sum())


negative_marks = (df[['math score','reading score','writing score']] < 0).sum().sum()

if negative_marks == 0:
    print("No Negative Marks Found")
else:
    print(f"Found {negative_marks} Negative Marks")

# Subject Averages
math_avg = df['math score'].mean()
reading_avg = df['reading score'].mean()
writing_avg = df['writing score'].mean()

print("\nAverages Scores")
print("Math:", round(math_avg,2))
print("Reading:", round(reading_avg,2))
print("Writing:", round(writing_avg,2))

# Average Score Per Student
df['Average Score'] = df[['math score','reading score','writing score']].mean(axis=1)

# Top 5 Students

top5 = df.nlargest(5, 'Average Score')
print("\nTop 5 Performing  Students")
print(top5)
```

```python
# Bar Chart
subjects = ['Math','Reading','Writing']
averages = [math_avg, reading_avg, writing_avg]

plt.figure(figsize=(6,4))
plt.bar(subjects, averages)
plt.title("Average Scores by Subject")
plt.ylabel("Average Marks")
plt.show()
```

```python
# Histogram
plt.figure(figsize=(8,5))
plt.hist(df['Average Score'], bins=15)
plt.title("Student Performance Distribution")
plt.xlabel("Average Score")
plt.ylabel("Number of Students")
plt.show()
```
